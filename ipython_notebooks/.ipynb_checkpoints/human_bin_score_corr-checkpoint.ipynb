```python
### parameters
cell_org = {'H1':'human',
            'GM':'human',
            'mCD8T:WT':'mouse',
            'mCD8T:DFMO':'mouse',
            'mCD8T:ODCKO':'mouse'}

cell_chrnames = {'H1':['chr%s' % (i) for i in range(1, 23)] + ['chrX', 'chrY'],
                 'GM':['chr%s' % (i) for i in range(1, 23)] + ['chrX'],
                 'mCD8T:WT':['chr%s' % (i) for i in range(1, 20)] + ['chrX'],
                 'mCD8T:DFMO':['chr%s' % (i) for i in range(1, 20)] + ['chrX'],
                 'mCD8T:ODCKO':['chr%s' % (i) for i in range(1, 20)] + ['chrX']}
```

```python
### exp list
### exp = (rep, cell, sample, agent, tnum)
#exp_list = [(1, 'H1', 'NCP', 'sp', 8),
#            (2, 'H1', 'NCP', 'sp', 8),
#            (1, 'H1', 'NCP', 'HP1a', 3),
#            (2, 'H1', 'NCP', 'HP1a', 3),
#            (1, 'H1', 'DNA', 'HP1a', 3),
#            (2, 'H1', 'DNA', 'HP1a', 3),
#            (1, 'GM', 'NCP', 'sp', 8),
#            (2, 'GM', 'NCP', 'sp', 8)]

exp_list = [(1, 'H1', 'NCP', 'sp', i) for i in range(1, 10)]
exp_list += [(2, 'H1', 'NCP', 'sp', i) for i in range(1, 10)]

deep = False
bin_size = 10000
dtype = 'score'
```

```python
### group exps over titrations
dexp_tnums = {}
for exp in exp_list:
    rep, cell, sample, agent, tnum = exp
    dexp = (rep, cell, sample, agent)
    if dexp not in dexp_tnums:
        dexp_tnums[dexp] = []
    dexp_tnums[dexp].append(tnum)
```

```python
### read data
exp_ID_score = {}
for dexp in dexp_tnums:
    rep, cell, sample, agent = dexp

    if deep:
        fname = '_'.join([cell,
                          sample,
                          agent,
                          str(rep) + 'rep',
                          'deep',
                          str(int(bin_size/1000.0)) + 'kb',
                          dtype]) + '.cn'

    else:
        fname = '_'.join([cell,
                          sample,
                          agent,
                          str(rep) + 'rep',
                          str(int(bin_size/1000.0)) + 'kb',
                          dtype]) + '.cn'
    
    field_names = []    
    for tnum in dexp_tnums[dexp]:
        if deep:
            field_name = '_'.join([cell,
                                   sample,
                                   agent,
                                   str(tnum),
                                   str(rep) + 'rep',
                                   'deep'])
        else:
            field_name = '_'.join([cell,
                                   sample,
                                   agent,
                                   str(tnum),
                                   str(rep) + 'rep'])

        field_names.append(field_name)

    chr_choices = cell_chrnames[cell]
        
    field_ID_value = load_file.read_cn_file(path + fname,
                                            mode='col',
                                            field_choices=field_names,
                                            chr_choices=chr_choices)

    for tnum in dexp_tnums[dexp]:
        exp = (rep, cell, sample, agent, tnum)

        if deep:
            field_name = '_'.join([cell,
                                   sample,
                                   agent,
                                   str(tnum),
                                   str(rep) + 'rep',
                                   'deep'])
        else:
            field_name = '_'.join([cell,
                                   sample,
                                   agent,
                                   str(tnum),
                                   str(rep) + 'rep'])

        assert exp not in exp_ID_score
        exp_ID_score[exp] = field_ID_value[field_name]
```

```python
### get common IDs
ID_list = set([])
for i in range(len(exp_list)):
    exp = exp_list[i]
    if i == 0:
        ID_list |= set(exp_ID_score[exp].keys())
        continue
    ID_list &= set(exp_ID_score[exp].keys())
ID_list = sorted(list(ID_list))
    
```

```python
### get correlation over tittration points
for i in range(1, 10):
    exp1 = (1, 'H1', 'NCP', 'sp', i)
    exp2 = (2, 'H1', 'NCP', 'sp', i)

    X, Y = [], []
    for ID in ID_list:
        X.append(exp_ID_score[exp1][ID])
        Y.append(exp_ID_score[exp2][ID])

    corr = scipy.stats.spearmanr(X, Y)[0]
    #corr = scipy.stats.pearsonr(X, Y)[0]
    print "%d-%s-%s-%s-%d" % exp1
    print "%d-%s-%s-%s-%d" % exp2
    print "%1.2f" % (corr)
    print 
    #print ("%s VS %s: %1.2f" % (agent1, agent2, corr))

    fig = plt.figure()
    plt.plot(X, Y, 'k,', alpha=0.5)
    plt.annotate("Spearman %1.2f" % (corr),
                 xy=(0.2, 0.75),
                 fontsize=12,
                 xycoords='axes fraction')
    #plt.title("%s VS %s" % (agent, agent2))
    plt.xlabel("fold change (%d-%s-%s-%s-%d)" % exp1)
    plt.ylabel("fold change (%d-%s-%s-%s-%d)" % exp2)
    #plt.xscale('log', base=2)
    #plt.yscale('log', base=2)
    plt.show()
    plt.close()


    
    
```

```python
### get correlation between experiments
pair_corr = {}
for i in range(len(exp_list)-1):
    for j in range(i+1, len(exp_list)):
        exp1 = exp_list[i]
        exp2 = exp_list[j]

        X, Y = [], []
        for ID in ID_list:
            X.append(exp_ID_score[exp1][ID])
            Y.append(exp_ID_score[exp2][ID])

        corr = scipy.stats.spearmanr(X, Y)[0]
        #corr = scipy.stats.pearsonr(X, Y)[0]
        print "%d-%s-%s-%s-%d" % exp1
        print "%d-%s-%s-%s-%d" % exp2
        print "%1.2f" % (corr)
        print 
        #print ("%s VS %s: %1.2f" % (agent1, agent2, corr))
            
        fig = plt.figure()
        plt.plot(X, Y, '.')
        plt.annotate("Spearman %1.2f" % (corr),
                     xy=(0.2, 0.75),
                     fontsize=12,
                     xycoords='axes fraction')
        #plt.title("%s VS %s" % (agent, agent2))
        plt.xlabel("fold change (%d-%s-%s-%s-%d)" % exp1)
        plt.ylabel("fold change (%d-%s-%s-%s-%d)" % exp2)
        #plt.xscale('log', base=2)
        #plt.yscale('log', base=2)
        #plt.show()
        plt.close()

        pair_corr[(exp1, exp2)] = corr
```

```python
### reorganize data and make labels
exp_label = {}
exp_data = {}
for exp in exp_list:
    rep, cell, sample, agent, tnum = exp
    label = '%s %s %s\nid:%d %d-rep' % (cell, sample, agent, tnum, rep)
    exp_label[exp] = label
    data = [exp_ID_score[exp][ID] for ID in ID_list]
    exp_data[exp] = data
    
```

```python
# plot correlation matrix between experiment
graphics.plot_corr_matrix(exp_data,
                          exp_label,
                          ids = exp_list,
                          scatter_style='dot',
                          bins=30,
                          cbar=True,
                          save=False,
                          title = "Correlation betwen experiments (1 kb bin)",
                          note='1kb')
```
