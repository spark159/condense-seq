```python
count = 0
for rlen_fnames, labels in zip([rlen_fnames0, rlen_fnames1, rlen_fnames2, rlen_fnames3, rlen_fnames4],
                               [labels0, labels1, labels2, labels3, labels4]):

    ### load rlen file
    rlen_counts = [load_file.read_rlen_file (rlen_path + fname) for fname in rlen_fnames]


    ### plot rlen distribution
    colors = np.linspace(0.01, 0.7, num=len(rlen_counts))
    alphas = np.linspace(0.3, 1, num=len(rlen_counts))[::-1]
    cmap = mpl.cm.get_cmap("copper")

    graphics.plot_rlen_dist (rlen_counts,
                             colors=[cmap(color) for color in colors],
                             alphas=alphas,
                             labels=labels,
                             fig_width=6,
                             fig_height=4,
                             note=str(count),
                             save=True,
                             save_path='data/')
    count+=1
```
